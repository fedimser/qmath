```python
# Evaluating log2(x), x in [1,2). 
# Reference: https://arxiv.org/pdf/2001.00807 (section 3.1.1)

import numpy as np

# r - precision
def log2_fbe(x, r=40):
    w=[0]*r
    assert 1<=x<2
    a=x**2
    for i in range(r):
        assert 1<=a<4
        if a>=2:
            w[i]=1
            a=a/2
        a=a**2
    return sum(w[i]*(2**(-(i+1))) for i in range(r))

for x in np.linspace(1, 1.99, 100):
    assert np.isclose(np.log2(x), log2_fbe(x))

print("OK")
```
