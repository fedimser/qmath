```python
# Evaluating 2^x, x in [0,1). 
# Reference: https://arxiv.org/pdf/2001.00807

import numpy as np

# Computes 2^(x * 2^-r).
def pow_qfbe(x, r):
    assert 0 <= x < 2**r
    v = [(x>>i)%2 for i in range(r)]
    a=1
    for i in range(r):
        if v[i]==1:
            a*=2
        a=a**0.5
    return a

r=15
for x in range(2**r):
    val1 = pow_qfbe(x, r)
    val2 = 2**(x * 2**(-r))
    assert np.isclose(val1, val2)
print("OK")
```
