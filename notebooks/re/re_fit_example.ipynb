A simple example of how to fit numeric resource esitimation with a polynomial.

```python
from qmath.utils.re_utils import re_numeric_int_binary_op
from qmath.add import CDKMAdder
import numpy as np

adder = CDKMAdder()
re_numeric = lambda assgn: re_numeric_int_binary_op(adder, assgn)

n_range = range(5,30)
re = [re_numeric({"n":n}) for n in n_range]
for metric in ["gidney_lelbows", "gidney_relbows", "toffs","qubit_highwater", "active_volume"]:
    print(metric, np.polyfit(n_range, [x[metric] for x in re], deg=3))
```
